```python
# Create a new database file
conn = sqlite3.connect('my_database.db')
cursor = conn.cursor()

# main process tables

# Final Upload Table 
cursor.execute('''CREATE TABLE upload_table (
                    upld_str_cd INTEGER,
                    upld_item_code INTEGER,
                    cap_units INTEGER,
                    delta_desc TEXT)''')
# Cap History 
cursor.execute('''CREATE TABLE cap_hist (
                    run_date DATE,
                    top_level_item INTEGER,
                    upld_item_code INTEGER,
                    store_nbr INTEGER,
                    upld_str_cd INTEGER,
                    delta_desc TEXT,
                    lgc_cd TEXT,
                    cap_units INTEGER)''')
# running 12 month cap details
cursor.execute('''CREATE TABLE cap_details (
                    run_date DATE,
                    top_level_item INTEGER,
                    store_nbr INTEGER,
                    ord_mult INTEGER,
                    budget_dos REAL,
                    str_dos REAL,
                    str_ftr_dos REAL,
                    oh_units INTEGER,
                    oh_dlrs REAL,
                    SLS_UNITS INTEGER,
                    CSU REAL,
                    arriv_units INTEGER,
                    arriv_cost REAL,
                    flex_days INTEGER,
                    pb_units INTEGER,
                    ONE_WOS_Units REAL,
                    TWO_WOS_Units REAL,
                    THREE_WOS_Units REAL,
                    FOUR_WOS_Units REAL,
                    ONE_FWOS_Units REAL,
                    TWO_FWOS_Units REAL,
                    THREE_FWOS_Units REAL,
                    FOUR_FWOS_Units REAL)''')
# Store On Hand Snapshot
cursor.execute('''CREATE TABLE str_oh (
                    run_date DATE,
                    top_level_item INTEGER,
                    store_nbr INTEGER,
                    oh_units INTEGER,
                    oh_dlrs REAL)''')
# Delivery Plan Table for Future Days of Supply
cursor.execute('''CREATE TABLE plan_deliv (
                    run_date DATE,
                    top_level_item INTEGER,
                    store_nbr INTEGER,
                    arrivdate DATE,
                    units INTEGER)''')
# Sales Table for Cost of Sales Units
cursor.execute('''CREATE TABLE sales_upc (
                    wk_end_dt DATE,
                    wk_per_cd INTEGER,
                    UPC TEXT,
                    SLS_UNITS INTEGER,
                    SLS_COST_DLRS REAL)''')
# SOURCING TABLES FOR Direct Delivery and Warehouse Association
cursor.execute('''CREATE TABLE sourcing (
                    item INTEGER,
                    str_nbr DATE,
                    source TEXT,
                    vend_mult INTEGER)''')
# DATE DIM TABLES
cursor.execute('''CREATE TABLE period_dim (
                    day_per_cd INTEGER,
                    wk_per_cd INTEGER,
                    DAY_DT DATE,
                    MONTH INTEGER,
                    YEAR INTEGER)''')
# TLI DIM TABLE FOR SPINE RELATION
cursor.execute('''CREATE TABLE DIM_TLI (
                    prod_cd INTEGER,
                    top_level_item INTEGER,
                    ob_ind TEXT,
                    item_dept TEXT,
                    item_brand TEXT)''')
# ITEM||STORE TABLE FOR STOCKED LENGTH
cursor.execute('''CREATE TABLE STR_IT_TBL (
                    store_nbr INTEGER,
                    prod_cd INTEGER,
                    per_cd INTEGER)''')
# Budget Table to associate DEPT budget to ITEM
cursor.execute('''CREATE TABLE budget (
                    wk_end_dt DATE,
                    wk_per_cd INTEGER,
                    top_level_item INTEGER,
                    item_dept INTEGER,
                    plan_dlrs REAL,
                    planned_csu REAL)''')

# utility process tables

# STR LAYOUT TABLE FOR COST COUNTS
cursor.execute('''CREATE TABLE STR_LAYOUT_TBL (
                    store_nbr INTEGER,
                    layout_cd INTEGER,
                    UPC TEXT)''')
# STR TABLE FOR COST COUNTS
cursor.execute('''CREATE TABLE STR_TBL (
                    store_nbr INTEGER,
                    layout_cd INTEGER)''')
# cost dim table
cursor.execute('''CREATE TABLE cost_dim (
                    whse_nbr INTEGER,
                    mid_level_item INTEGER,
                    vend_num INTEGER,
                    WHS_CST REAL,
                    eff_frm DATE,
                    run_date DATE)''')
# casepack table
cursor.execute('''CREATE TABLE casepack (
                    whse_nbr INTEGER,
                    mid_level_item INTEGER,
                    vend_num INTEGER,
                    CASEPACK_QTY INTEGER,
                    INPK_QTY INTEGER,
                    eff_frm DATE,
                    run_date DATE)''')
# Order Multiple table
cursor.execute('''CREATE TABLE Ord_Mult (
                    whse_nbr INTEGER,
                    mid_level_item INTEGER,
                    ord_mult INTEGER,
                    eff_date DATE,
                    run_date DATE)''')
# MLI table
cursor.execute('''CREATE TABLE dim_mid_level_item (
                    UPC TEXT,
                    mid_level_item INTEGER,
                    top_level_item INTEGER)''')
# vndr_it_dim
cursor.execute('''CREATE TABLE vndr_it_dim (
                    whse_nbr INTEGER,
                    mid_level_item INTEGER,
                    vend_num INTEGER,
                    eff_frm INTEGER,
                    run_date DATE)''')
# Final Utility Table
#cursor.execute('''CREATE TABLE TLI_DC_DATA (
#                    top_level_item INTEGER,
#                    whse_nbr INTEGER,
#                    Ord_Mult INTEGER,
#                    CASEPACK_QTY INTEGER,
#                    INPK_QTY INTEGER,
#                    whs_cst REAL)''')
# Commit and close
conn.commit()
conn.close()
```

```python
#utility builder
conn = sqlite3.connect('my_database.db')
cursor = conn.cursor()
om_query = """
CREATE TEMPORARY TABLE temp_om AS
with 
  maxdate as 
    (select max(run_date) from Ord_Mult)
select distinct
    all_dt_item.Top_Level_Item,
    all_dt_item.whse_nbr,
    all_dt_item.Ord_Mult
from
  (select distinct
    TLI.Top_Level_Item
    ,OM.Mid_Level_Item
    ,OM.whse_nbr
    ,OM.eff_date as OM_EFF_DT
    ,OM.Ord_Mult
  from Ord_Mult OM
  join dim_mid_level_item TLI
    on OM.Mid_Level_Item = TLI.mid_level_item
  where 1=1 
    and OM.run_date = (select * from maxdate)
    and OM.eff_date =
      (select max(eff_date) 
      from Ord_Mult a
      where OM.Mid_Level_Item = a.Mid_Level_Item 
        and OM.whse_nbr = a.whse_nbr)
  ) all_dt_item
join 
  (select distinct
    TLI.Top_Level_Item
    ,a.whse_nbr
    ,max(a.eff_date) as max_OM_EFF_DT
  from Ord_Mult a
  join dim_mid_level_item TLI
    on a.Mid_Level_Item = TLI.mid_level_item
  where 1=1 
    and a.run_date = (select * from maxdate)
    and a.eff_date = 
      (select max(eff_date) 
      from Ord_Mult c
      where a.Mid_Level_Item = c.Mid_Level_Item 
        and a.whse_nbr = c.whse_nbr)
    group by
      TLI.Top_Level_Item
      ,a.whse_nbr 
  ) eff_dt_item
on all_dt_item.Top_Level_Item = eff_dt_item.Top_Level_Item
  and all_dt_item.whse_nbr = eff_dt_item.whse_nbr
  and all_dt_item.OM_EFF_DT = eff_dt_item.max_OM_EFF_DT
"""
#doop query
doop_query = """
CREATE TEMPORARY TABLE om_doop AS
select distinct 
  a.Top_Level_Item, 
  a.whse_nbr, 
  max(a.Ord_Mult) as Ord_Mult
from temp_om a
join 
  (select distinct Top_Level_Item, whse_nbr
  from temp_om
  group by Top_Level_Item, whse_nbr
  having count(*)>1) b
on a.Top_Level_Item = b.Top_Level_Item 
  and a.whse_nbr = b.whse_nbr
group by a.Top_Level_Item, a.whse_nbr
order by a.Top_Level_Item, a.whse_nbr"""
#doop delete
delete_query="""
CREATE TEMPORARY TABLE adj_temp_om AS
select a.*
from temp_om a
--if not using sqlite i'd do an anti join
left join om_doop b
on a.Top_Level_Item = b.Top_Level_Item 
  and a.whse_nbr = b.whse_nbr
where b.Top_Level_Item IS NULL
  and b.whse_nbr IS NULL"""
insert_query="""
CREATE TABLE ORDER_MULT AS 
select * from adj_temp_om
union 
select * from om_doop
order by Top_Level_Item, whse_nbr
"""
cursor.execute(om_query)
cursor.execute(doop_query)
cursor.execute(delete_query)
cursor.execute(insert_query)
#df = pd.read_sql_query("select * from temp_om", conn)
#df = pd.read_sql_query(om_query, conn)
#display(df) 

conn.commit()
conn.close()
```

```python
conn = sqlite3.connect('my_database.db')
cursor = conn.cursor()

cspk_query ="""
CREATE TEMPORARY TABLE temp_cspk AS
with 
  maxdate as
    (select max(run_date) from casepack),
  maxdate2 as
    (select max(run_date) from vndr_it_dim)
select distinct
  all_dt_item.Top_Level_Item,
  all_dt_item.whse_nbr,
  all_dt_item.CASEPACK_QTY,
  all_dt_item.INPK_QTY
from
  (select distinct
    TLI.Top_Level_Item
    ,casepack.Mid_Level_Item
    ,vend.vend_num
    ,casepack.whse_nbr
    ,casepack.EFF_FRM as CASEPACK_EFF
    ,vend.eff_frm as vend_eff
    ,CASEPACK_QTY
    ,INPK_QTY
  from casepack
  join vndr_it_dim vend
    on casepack.vend_num = vend.vend_num
    and casepack.Mid_Level_Item = vend.Mid_Level_Item
    and casepack.whse_nbr = vend.whse_nbr
  join dim_mid_level_item TLI
    on casepack.Mid_Level_Item = TLI.mid_level_item
  where casepack.run_date = (select * from maxdate)
    and casepack.EFF_FRM = 
      (select max(a.EFF_FRM) 
      from casepack a
      where a.run_date = (select * from maxdate)
        and casepack.Mid_Level_Item = a.Mid_Level_Item 
        and casepack.vend_num = a.vend_num
        and casepack.whse_nbr = a.whse_nbr)
    and vend.eff_frm =
      (select max(b.eff_frm) 
      from vndr_it_dim b
      where b.run_date = (select * from maxdate2)
        and vend.Mid_Level_Item = b.Mid_Level_Item 
        and vend.vend_num = b.vend_num
        and vend.whse_nbr = b.whse_nbr)
  ) all_dt_item
join 
  (select distinct
    TLI.Top_Level_Item
    ,casepack.whse_nbr
    ,max(casepack.EFF_FRM) as max_case_eff
    ,max(vend.eff_frm) as max_vendor_eff
  from casepack 
  join vndr_it_dim vend
    on casepack.vend_num = vend.vend_num
    and casepack.Mid_Level_Item = vend.Mid_Level_Item
    and casepack.whse_nbr = vend.whse_nbr
  join dim_mid_level_item TLI
    on casepack.Mid_Level_Item = TLI.mid_level_item
  where casepack.run_date = (select * from maxdate)
    and casepack.EFF_FRM = 
      (select max(a.EFF_FRM) 
      from casepack a
      where a.run_date = (select * from maxdate)
        and casepack.Mid_Level_Item = a.Mid_Level_Item 
        and casepack.vend_num = a.vend_num
        and casepack.whse_nbr = a.whse_nbr)
    and vend.eff_frm =
      (select max(b.eff_frm)
      from vndr_it_dim b
      where b.run_date = (select * from maxdate2)
        and vend.Mid_Level_Item = b.Mid_Level_Item 
        and vend.vend_num = b.vend_num
        and vend.whse_nbr = b.whse_nbr)
    group by
      TLI.Top_Level_Item
      ,casepack.whse_nbr 
  ) eff_dt_item
on all_dt_item.Top_Level_Item = eff_dt_item.Top_Level_Item
  and all_dt_item.whse_nbr = eff_dt_item.whse_nbr
  and all_dt_item.CASEPACK_EFF = eff_dt_item.max_case_eff
  and all_dt_item.vend_eff = eff_dt_item.max_vendor_eff
order by all_dt_item.Top_Level_Item, all_dt_item.whse_nbr
"""
doop_query="""
CREATE TEMPORARY TABLE cspk_doop AS
select distinct
  cs.Top_Level_Item,
  cs.whse_nbr,
  cs.CASEPACK_QTY,
  cs.INPK_QTY
from temp_cspk cs
join
  (select distinct 
    a.Top_Level_Item,
    a.whse_nbr,
    max(a.CASEPACK_QTY+a.INPK_QTY) as pacmax
  from temp_cspk a
  join 
    (select distinct Top_Level_Item, whse_nbr
    from temp_cspk
    group by Top_Level_Item, whse_nbr 
    having count(*) > 1) b 
  on a.Top_Level_Item = b.Top_Level_Item
    and a.whse_nbr = b.whse_nbr
  group by 
    a.Top_Level_Item,
    a.whse_nbr) dup
on cs.Top_Level_Item = dup.Top_Level_Item
  and cs.whse_nbr = dup.whse_nbr
  and (cs.CASEPACK_QTY+cs.INPK_QTY) = dup.pacmax"""
#doop delete
delete_query="""
CREATE TEMPORARY TABLE adj_temp_cspk AS
select a.*
from temp_cspk a
--if not using sqlite i'd do an anti join
left join cspk_doop b
on a.Top_Level_Item = b.Top_Level_Item 
  and a.whse_nbr = b.whse_nbr
where b.Top_Level_Item IS NULL
  and b.whse_nbr IS NULL"""
insert_query="""
CREATE TABLE CSPK AS 
select * from adj_temp_cspk
union 
select * from cspk_doop
order by Top_Level_Item, whse_nbr
"""
cursor.execute(cspk_query)
cursor.execute(doop_query)
cursor.execute(delete_query)
cursor.execute(insert_query)

conn.commit()
conn.close()
```

```python
conn = sqlite3.connect('my_database.db')
cursor = conn.cursor()

cost_query ="""
CREATE TEMPORARY TABLE temp_cost AS
select distinct
  all_dt_item.whse_nbr,
  all_dt_item.Top_Level_Item,
  all_dt_item.Mid_Level_Item,
  all_dt_item.whs_cst
from 
  (select distinct
    cst.whse_nbr,
    cst.Mid_Level_Item,
    TLI.Top_Level_Item,
    cst.whs_cst,
    cst.EFF_FRM as EFF_DATE
  from cost_dim cst
  join vndr_it_dim vend
    on cst.vend_num = vend.vend_num
    and cst.Mid_Level_Item = vend.Mid_Level_Item
    and cst.whse_nbr = vend.whse_nbr
  join dim_mid_level_item TLI 
    on cst.Mid_Level_Item = TLI.MID_LEVEL_ITEM
  where cst.run_date = (select max(run_date) from cost_dim)
    and cst.EFF_FRM = 
      (select max(a.EFF_FRM) 
      from cost_dim a
      where a.run_date = (select max(run_date) from cost_dim)
        and cst.Mid_Level_Item = a.Mid_Level_Item 
        and cst.whse_nbr = a.whse_nbr
        and cst.vend_num = a.vend_num)
    and vend.eff_frm =
      (select max(eff_frm) 
      from vndr_it_dim b
      where b.run_date = (select max(run_date) from vndr_it_dim)
        and vend.Mid_Level_Item = b.Mid_Level_Item 
        and vend.vend_num = b.vend_num
        and vend.whse_nbr = b.whse_nbr)
  ) all_dt_item
join 
  (select distinct
    cst.whse_nbr,
    TLI.Top_Level_Item,
    max(cst.EFF_FRM) as MAX_EFF_DATE
  from cost_dim cst
  join vndr_it_dim vend
    on cst.vend_num = vend.vend_num
    and cst.Mid_Level_Item = vend.Mid_Level_Item
    and cst.whse_nbr = vend.whse_nbr
  join dim_mid_level_item TLI 
    on cst.Mid_Level_Item = TLI.MID_LEVEL_ITEM
  where cst.run_date = (select max(run_date) from cost_dim)
    and cst.EFF_FRM = 
      (select max(a.EFF_FRM) 
      from cost_dim a
      where a.run_date = (select max(run_date) from cost_dim)
        and cst.Mid_Level_Item = a.Mid_Level_Item 
        and cst.whse_nbr = a.whse_nbr
        and cst.vend_num = a.vend_num)
    and vend.eff_frm =
      (select max(eff_frm) 
      from vndr_it_dim b
      where b.run_date = (select max(run_date) from vndr_it_dim)
        and vend.Mid_Level_Item = b.Mid_Level_Item 
        and vend.vend_num = b.vend_num
        and vend.whse_nbr = b.whse_nbr)
  group by 
    cst.whse_nbr, 
    TLI.Top_Level_Item
  ) eff_dt_item
on all_dt_item.Top_Level_Item = eff_dt_item.Top_Level_Item
  and all_dt_item.whse_nbr = eff_dt_item.whse_nbr
  and all_dt_item.EFF_DATE = eff_dt_item.MAX_EFF_DATE
order by all_dt_item.Top_Level_Item
"""
doop_query="""
CREATE TEMPORARY TABLE cost_doop AS
  select distinct
  a.Top_Level_Item,
  a.whse_nbr,
  max(a.WHS_CST) as WHS_CST
  from temp_cost a
  join 
    (Select distinct 
      Top_Level_Item, 
      whse_nbr
    from temp_cost
    group by 
      Top_Level_Item,
      whse_nbr
    having count(*) > 1) b
  on a.Top_Level_Item = b.Top_Level_Item
    and a.whse_nbr = b.whse_nbr
  group by
  a.Top_Level_Item
  ,a.whse_nbr
"""
delete_query="""
CREATE TEMPORARY TABLE adj_temp_cost AS
select a.*
from temp_cost a
--if not using sqlite i'd do an anti join
left join cost_doop b
on a.Top_Level_Item = b.Top_Level_Item 
  and a.whse_nbr = b.whse_nbr
where b.Top_Level_Item IS NULL
  and b.whse_nbr IS NULL"""
insert_query="""
CREATE TABLE COST AS
select 
  Top_Level_Item,
  whse_nbr,
  whs_cst 
from adj_temp_cost
union 
select * from cost_doop 
order by Top_Level_Item, whse_nbr
"""
cursor.execute(cost_query)
cursor.execute(doop_query)
cursor.execute(delete_query)
cursor.execute(insert_query)

conn.commit()
conn.close()
```

```python
conn = sqlite3.connect('my_database.db')
cursor = conn.cursor()

finjoin_query="""
CREATE TABLE TLI_DC_DATA AS
select distinct 
  spine.Top_Level_Item,
  spine.whse_nbr,
  OM.Ord_Mult,
  cspk.CASEPACK_QTY,
  cspk.INPK_QTY,
  COST.WHS_CST
from 
(select distinct Top_Level_Item, whse_nbr from
  (select distinct Top_Level_Item, whse_nbr from ORDER_MULT
  union
  select distinct Top_Level_Item, whse_nbr from CSPK
  union
  select distinct Top_Level_Item, whse_nbr from COST)) spine
left join ORDER_MULT OM 
  on spine.Top_Level_Item = OM.Top_Level_Item 
  and spine.whse_nbr = OM.whse_nbr
left join CSPK 
  on spine.Top_Level_Item = CSPK.Top_Level_Item 
  and spine.whse_nbr = CSPK.whse_nbr
left join COST   
  on spine.Top_Level_Item = COST.Top_Level_Item 
  and spine.whse_nbr = COST.whse_nbr
"""
cursor.execute(finjoin_query)
conn.commit()
conn.close()
```
